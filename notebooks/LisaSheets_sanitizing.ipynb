```python
import pandas as pd
import re
```

```python
df = pd.read_csv('dataset.csv')
```

```python
df = df.drop('question', axis=1)
```

```python
def remove_double_bracketed_text(text):
    # Use regex to find and remove text within {{...}}
    text_cleaned_of_braces = re.sub(r"\{\{.*?\}\}", "", text, flags=re.DOTALL).strip()
    # text_cleaned_of_square_brackets = re.sub(r"\[\[.*?\]\]", "", text_cleaned_of_braces, flags=re.DOTALL).strip()
    return text_cleaned_of_braces

def extract_id_from_content(content):
    # Updated regex to match the new pattern
    match = re.search(r"\|Identifiant=(.*?)\s*\|", content)
    return match.group(1) if match else None
```

```python
df['id'] = df['content'].apply(extract_id_from_content)
df.head()
```

```python
df['lisa_sheet'] = df.content.apply(remove_double_bracketed_text)
df.head()
```

```python
df.to_csv('lisa_sheets.csv')
```
