```python
df = pd.read_csv('dataset_dpo_v2.csv')
df.head()
```

```python
df['label'] = df['quality_score'] == '5'
```

```python
df = df.drop(['source', 'quality_score'], axis=1)
df.head()
```

```python
df['completion'] = df.apply(
    lambda row: (
        f"Question: {row['question']}\n"
        f"a) {row['option_a']}\n"
        f"b) {row['option_b']}\n"
        f"c) {row['option_c']}\n"
        f"d) {row['option_d']}"
    ),
    axis=1
)
```

```python
df = df.drop(['option_a', 'option_b', 'option_c', 'option_d'], axis=1)
df.head()
```

```python
df = df[['question', 'completion', 'label']]
```

```python
df['label'] = df['label'].astype(bool)
df.head()
```

```python
df.to_csv('kto_dataset_formatted.csv')
```
